```python
country_uses_python  = country_group['LanguageWantToWorkWith'].apply(lambda x : x.str.contains('Python').sum())
country_uses_python 
```

```python
country_rsp = df['Country'].value_counts()
country_rsp.head()
```

```python
python_df = pd.concat([country_rsp,country_uses_python ],axis='columns',sort=False)
python_df
```

```python
python_df.rename(columns={'Country':'NumOfRespondants','LanguageWantToWorkWith':'NumWorkedWithPython'},inplace=True)
python_df.head(5)
```

```python
python_df['PctKnowsPython'] = (python_df['NumWorkedWithPython']/python_df['NumOfRespondants'])*100
python_df.head(10)
```

```python
python_df.sort_values(by='PctKnowsPython',ascending=False,inplace=True)
python_df.head(50)
```
