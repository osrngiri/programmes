```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.histplot(data=tips,x='total_bill')
```

```python
from matplotlib import pyplot as plt
sns.scatterplot(x='total_bill',y='tip', data=tips,hue='sex',size = 'size',sizes=(20,50))
plt.xlabel('Total Bill')
plt.ylabel('Tips')
plt.title('Relation between Total Bill and Tips')
```

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event',style='region',markers=True,dashes=False)
plt.xlabel("Timepoint")
plt.ylabel("Signal Intensity")
plt.title("Changes in Signal Intensity over Time")
```

```python
ex = sns.load_dataset('exercise')
ex.head()
```

```python
sns.barplot(x='kind',y='pulse',hue='diet',errorbar=None, palette="muted",data=ex)
plt.xlabel("Activity")
plt.ylabel("Pulse Rate")
plt.title("Variation of Pulse with Activity")
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.histplot(data=iris,x='petal_length',bins=20,kde=True,color='green')
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.title("Distribution of Petal Lengths in Iris Flowers")
```

```python
sns.kdeplot(data=tips,x='total_bill',hue='time',fill=True,alpha=0.6,linewidth=0.5)
plt.title("Density Plot of Total Bill by Meal Time")
plt.xlabel("Total Bill ($)")
plt.ylabel("Density")
```

```python
sns.boxplot(x = 'day',y = 'total_bill',data=tips,hue='time',palette='Set3',linewidth=1.5,fliersize=4)
plt.title("Box Plot of Total Bill by Day and Meal Time")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill ($)")
```

```python
plt.figure(figsize=(8,6))
sns.violinplot(x = 'species',y = 'petal_length',data=iris)
```

```python
corr = tips.corr(numeric_only=True)
sns.heatmap(corr)
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights = flights.pivot('month', 'year', 'passengers')
sns.heatmap(flights,cmap='Blues',annot=True,fmt='d')
plt.title('Passengers per month')
plt.xlabel('Year')
plt.ylabel('Month')
```

```python
sns.pairplot(data = iris)
```

```python
sns.pairplot(data = iris,hue = 'species', diag_kind='kde',palette='husl')
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris)
```

```python
g = sns.FacetGrid(tips, col="day")
g.map(sns.histplot,'total_bill')
```

```python
sns.scatterplot(x='total_bill', y='tip',hue='day',data=tips, palette='Set2')
plt.title("Total Bill vs Tip")
plt.xlabel('Total Bill($)')
plt.ylabel('Tip($)')
```
